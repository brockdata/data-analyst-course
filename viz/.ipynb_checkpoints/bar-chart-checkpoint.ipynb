```python
import pandas as pd
import matplotlib.pyplot as plt
# seaborn is built on top of matplotlib
import seaborn as sns
# overwrite matplotlib "look" with seaborn
sns.set()
```

```python
raw_data = pd.read_csv('bar-chart-data.csv')
raw_data
df = raw_data.copy()
```

```python
# increase size
plt.figure(figsize=(10,10))
# create plot
plt.bar(x=df['Brand'],
       height=df['Cars Listings'],
       color='midnightblue')
# rotate labels
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)
# title
plt.title('Cars Listings by Brand', fontsize=16, fontweight='bold')
# label
plt.ylabel('Number of Listings', fontsize=13)

# creates a new figure when showing
# SAVE BEFORE THIS
# save
plt.savefig('bar.png')

plt.show()

# color options
# letter e.g. ['r','g','b','w','y','m','c']
# name 'midnightblue'
```
