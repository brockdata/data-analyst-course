```python
# time series
# e.g. stock market
# S&P500 = 500 largest US companies
# FTSE100 = largest 100 UK
```

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('line-chart-data.csv')
raw_data.head()
```

```python
df = raw_data.copy()
```

```python
# first, convert to datetime
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'].info())
print(df.head())
```

```python
# slice post 2008
df = df.loc[(df['Date'] >= dt.datetime(year=2008, month=1, day=7)) & 
            (df['Date'] <= dt.datetime(year=2008, month=12, day=31))]
print(df['Date'].min())
print(df['Date'].max())
```

```python
labels = ['S&P500', 'FTSE100']

plt.figure(figsize=(20,8))
plt.plot(df['Date'],
        df['GSPC500'],
        color='midnightblue')
plt.plot(df['Date'],
        df['FTSE100'],
        color='crimson')
plt.title('S&P vs FTSE (Q3 & Q4 2008)', fontsize=14, fontweight='bold')
plt.ylabel('Returns')
plt.xlabel('Date')
plt.legend(labels=labels, fontsize='large')

plt.show()
```
