```python
# regressions
# relationship between dependent/outcome and independent/predictor
# line of best fit closest to all points
# linear = flat trend line
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('regression-data.csv')
raw_data.head()
```

```python
df = raw_data.copy()
```

```python
# use seaborn
# regplot
# accepts data in df, numpy array, etc.

# 90% confidence interval added by default

# matplotlib size
# plt.figure(figsize=(10,8))

# seaborn size
sns.set(rc={'figure.figsize': (9,6)})

# add "color=" here for line and  points
# add "scatter_kws=" and "line_kws=" for different color for line
sns.regplot(x='Budget',
           y='Sales',
           data=df,
            scatter_kws={'color': 'k'},
            line_kws={'color': 'red'})
plt.xlabel('Ad Expenditure ($Thousands)')
plt.ylabel('Sales (Unit Thousands)', fontsize=14, fontweight='bold')


plt.show()
```

```python
# lmplot
# accepts other data formats beyond normal
sns.lmplot(x='Budget',
          y='Sales',
          data=df,
          height=10,
            scatter_kws={'color': 'k'},
            line_kws={'color': 'red'})
plt.xlabel('Ad Expenditure ($Thousands)')
plt.ylabel('Sales (Unit Thousands)', fontsize=14, fontweight='bold')


plt.show()
```

```python
# interpretation

# relationship between two variables
# specificity -- relationship and specific change
# i.e. parameter on the trendline: 100k additional ad expenditure = 5k additonal units sold

# choosing x and y
# y = f(x) --  get the axes correctly

# different software >> different trendlines

# R2 varies with the industry
```
