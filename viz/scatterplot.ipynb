```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('scatter-data.csv')
raw_data.head()
```

```python
df = raw_data.copy()
```

```python
# ----- matplotlib
plt.figure(figsize=(12,8))
# alpha = transparency / density
# cmap 'mako' or 'viridis' color schemes
# store in variable to use the pointer below
scatter = plt.scatter(df[r'Area (ft.)'],
           df['Price'],
           alpha=0.6,
           c=df['Building Type'],
           cmap='viridis')
# pointer --> memory cell where first element of the scatter's "legend_elements" is
plt.legend(*scatter.legend_elements(),
          loc='upper left',
          title='Building Type')
plt.title('Relationship Between Area and Price of California Real Estate',
          fontsize=14, fontweight='bold')
plt.xlabel(r'Area (sqft)', weight='bold')
plt.ylabel(r'Price($Thousands)')

plt.show()
```

```python
# ----- seaborn
plt.figure(figsize=(12,8))
sns.scatterplot(x=df['Price'],
               y=df[r'Area (ft.)'],
               hue=df['Building Type'],
               palette=['black', 'darkblue', 'purple', 'pink', 'white'],
               s=100)
# paste from matplotlib above... same code
plt.title('Relationship Between Area and Price of California Real Estate',
          fontsize=14, fontweight='bold')
plt.xlabel(r'Area (sqft)', weight='bold')
plt.ylabel(r'Price($Thousands)')

plt.show()
```
