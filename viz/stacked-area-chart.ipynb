```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('stacked-area-chart-data.csv')
raw_data.head()
```

```python
df = raw_data.copy()
```

```python
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Diesel', 'Petrol', 'Gas']

# background
sns.set_style('white')

# figure size
plt.figure(figsize=(12,6))
# figure
plt.stackplot(df['Year'],
             df['Diesel'],
             df['Petrol'],
             df['Gas'],
             colors=colors,
             edgecolor='none')
# legend using "labels" list
plt.legend(labels=labels, loc='upper left')
plt.xticks(df['Year'], rotation=45)
plt.ylabel('Number of Cars',fontsize=13)
plt.title('Popularity of Fuel Types (1982 - 2016)', fontsize=14, weight='bold')
# remove borders
sns.despine()

plt.show()
```

```python
# interpretation
# spot trends for several categories

# petrol and diesel
# sample mostly cars in the early 2000s

# 2008 - used car purchases dipped

# diesel has dominated
```

```python
# when to use

# time data
# compare volume among categories (at least 3)
# two are ordered, numerical (e.g. year, num cars)
```

```python
# when NOT to use
# exact size is ambigious... consider a line
# otherwise, order by size (biggest on the bottom) and remove unneccessary categories
```
